```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np
import cv2
```

```python
model = load_model("forest_trial.h5")
```

```python
video = cv2.VideoCapture(0)  # Here try to put and see 0 ---> live video streaming
is_frame = True

while is_frame:
    is_frame,frame = video.read()  # read() returns 2 values, the image frame and boolean value if img present
    if is_frame:
        print(frame.shape)   # The default frame shape for live streming is (480, 640, 3)
        y = cv2.resize(frame,(64,64))
        x = np.expand_dims(y,axis=0)
        pred_prob = model.predict(x)
        if(pred_prob[0]==0):
            print('No fire is Detected')
        else:
            print('Fire is Detected !')
        cv2.imshow("Frame", frame)
        if cv2.waitKey(25) & 0xFF == ord('q'): # ord('q') --> ascii value of 'q' --> 113
            break
    else:
        break
cv2.destroyAllWindows()
video.release()
```
